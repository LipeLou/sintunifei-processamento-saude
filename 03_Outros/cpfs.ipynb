```python
import pandas as pd
```

```python
dados = pd.read_excel('/content/DADOS PARA A DIRF - trabalhando.xlsx', sheet_name='AGO 2025', index_col=False, skiprows=1)
cpfs = pd.read_excel('/content/DADOS PARA A DIRF - trabalhando.xlsx', sheet_name='CPFs faltando', index_col=False)

for index_cpfs, row_cpfs in cpfs.iterrows():
    nome_cpfs = row_cpfs['NOME']
    dependencia_cpfs = row_cpfs['DEPENDENCIA']
    cpf_cpfs = row_cpfs['CPF']

    # Find rows in 'dados' where 'NOME' matches and 'DEPENDENCIA' and 'CPF' are NaN
    matching_rows_indices = dados[(dados['NOME'] == nome_cpfs) & dados['DEPENDENCIA'].isna() & dados['CPF'].isna()].index

    # Update the matching rows in 'dados'
    for index_dados in matching_rows_indices:
        if pd.isna(dados.loc[index_dados, 'DEPENDENCIA']):
            dados.loc[index_dados, 'DEPENDENCIA'] = dependencia_cpfs
        if pd.isna(dados.loc[index_dados, 'CPF']):
            dados.loc[index_dados, 'CPF'] = cpf_cpfs

# Display the updated 'dados' DataFrame to verify the changes
dados.to_excel('AGOSTO.xlsx', index=0)
```
