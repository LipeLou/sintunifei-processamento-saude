```python
import pandas as pd
```

```python
emails = pd.read_csv('/content/planilha_sindicalizado_1.csv', index_col=0, encoding='latin1', sep=";")
emails.head()
```

```python
dados = pd.read_excel('/content/Planilha sem t√≠tulo.xlsx', index_col=0)
dados.head()
```

```python
for index, row in emails.iterrows():
    name = index
    email_principal = row['E-MAIL PRINCIPAL']
    outro_email = row['OUTRO E-MAIL']

    if name in dados.index:
        if pd.isna(dados.loc[name, 'Email']) and pd.notna(email_principal):
            dados.loc[name, 'Email'] = email_principal
        elif pd.isna(dados.loc[name, 'Email']) and pd.notna(outro_email):
            dados.loc[name, 'Email'] = outro_email

display(dados.head())
```

```python
dados.to_excel('dados.xlsx', index=0)
```
