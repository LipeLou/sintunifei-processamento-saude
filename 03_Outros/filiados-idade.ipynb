```python
import pandas as pd
```

```python
colunas_selecionadas = ["CARTEIRA", "DEPENDENCIA", "NASCIMENTO", "SEXO", "ESTADO CIVIL", "CÓD MÓDULO", "MÓDULO", "INICIO VIGÊNCIA"]
dados = pd.read_excel('/content/drive/MyDrive/filiados.xlsx', usecols=colunas_selecionadas)
```

```python
dados.head()
```

```python
dados['NASCIMENTO'] = pd.to_datetime(dados['NASCIMENTO'], format='%d/%m/%Y')
dados.head()
```

```python
data_referencia = pd.to_datetime('12/11/2025', format='%d/%m/%Y')
dados['IDADE'] = ((data_referencia - dados['NASCIMENTO']).dt.days / 365.25).astype(int)
dados.head()
```

```python
dados['MÓDULO'] = dados['MÓDULO'].str.replace('AMB HOSP OBST ', '', regex=False)
dados.head()
```

```python
pricing_rules = {
    'ENF': [
        (18, 256.90),
        (23, 363.70),
        (28, 417.11),
        (33, 567.80),
        (38, 567.80),
        (43, 567.80),
        (48, 597.47),
        (53, 597.47),
        (58, 638.56),
        (float('inf'), 808.86)
    ],
    'APT': [
        (18, 336.54),
        (23, 476.36),
        (28, 546.34),
        (33, 743.78),
        (38, 743.78),
        (43, 743.78),
        (48, 782.70),
        (53, 782.70),
        (58, 836.55),
        (float('inf'), 925.66)
    ]
}
print("Pricing rules dictionary updated successfully.")
```

```python
def calculate_valor(row):
    idade = row['IDADE']
    modulo = row['MÓDULO']

    if modulo in pricing_rules:
        for age_limit, valor in pricing_rules[modulo]:
            if idade <= age_limit:
                return valor
    return 0.00 # Default value if no rule matches or module not found

dados['VALOR'] = dados.apply(calculate_valor, axis=1)
dados.head()
```

```python
dados.to_excel('dados-quantitativos.xlsx', index=0)
```

# Task
Categorize the 'IDADE' column from the `dados` DataFrame into predefined age groups, then calculate and display the percentage of individuals falling into each age group.

## Define age bins and labels

### Subtask:
Define the age ranges as bins and create corresponding labels for the age groups.


**Reasoning**:
The subtask requires defining age bins and corresponding labels based on the provided pricing rules. This step creates the necessary lists for binning the 'IDADE' column.



```python
age_bins = [0, 18, 23, 28, 33, 38, 43, 48, 53, 58, float('inf')]
age_labels = ['0-18', '19-23', '24-28', '29-33', '34-38', '39-43', '44-48', '49-53', '54-58', '59+']
print("Age bins and labels defined successfully.")
```

## Categorize ages into groups

### Subtask:
Use `pd.cut` to categorize the 'IDADE' column in the `dados` DataFrame into the defined age groups.


**Reasoning**:
To categorize the 'IDADE' column into predefined age groups, I will use `pd.cut` with the previously defined `age_bins` and `age_labels` and store the result in a new column called 'GRUPO_IDADE'.



```python
dados['GRUPO_IDADE'] = pd.cut(dados['IDADE'], bins=age_bins, labels=age_labels, right=False)
dados.head()
```

**Reasoning**:
Now that the 'GRUPO_IDADE' column has been created, I will calculate the percentage of individuals in each age group by using `value_counts(normalize=True)` on this column, multiplying by 100 for percentages, and then displaying the results.



```python
age_group_percentage = dados['GRUPO_IDADE'].value_counts(normalize=True) * 100
print("Percentage of individuals in each age group:\n", age_group_percentage.round(2))
```
