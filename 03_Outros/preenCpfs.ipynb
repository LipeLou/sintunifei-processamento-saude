```python
import pandas as pd
```

```python
cpfs = pd.read_excel('/content/DADOS PARA A DIRF - trabalhando.xlsx', sheet_name='CPFs faltando', index_col=False)
dados = pd.read_excel('/content/DADOS PARA A DIRF - trabalhando.xlsx', sheet_name='MAI 2025', index_col=False, skiprows=1)
```

```python
dados.head()
```

```python
cpfs.head()
```

```python
for index, row in cpfs.iterrows():
    name_to_find = row['NOME']
    cpf_value = row['CPF']
    dependencia_value = row['DEPENDENCIA']

    # Find the corresponding row in the 'dados' DataFrame
    matching_row_index = dados[dados['NOME'] == name_to_find].index

    # If a matching row is found, update the 'CPF' and 'DEPENDENCIA' columns
    if not matching_row_index.empty:
        dados.loc[matching_row_index, 'CPF'] = cpf_value
        dados.loc[matching_row_index, 'DEPENDENCIA'] = dependencia_value

display(dados.head())
```

```python
null_cpfs_count = dados[dados['NOME'].notnull()]['CPF'].isnull().sum()
print(f"Number of null CPF values in 'dados' where NOME is not null: {null_cpfs_count}")
```

```python
dados.to_excel('cpfs_mai.xlsx', index=False)
```
